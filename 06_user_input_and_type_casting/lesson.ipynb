## 6.2 The input() Function

The **`input()`** function pauses your program and waits for the user to type something.

### Basic Syntax
```python
variable = input(prompt_message)
```

- **prompt_message**: The text shown to the user
- **Returns**: Whatever the user types (always as a string)

### First Example

````python
# Basic input example
name = input("What is your name? ")
print("Hello,", name, "!")
````

````python
# Multiple inputs
first_name = input("Enter your first name: ")
last_name = input("Enter your last name: ")
print("Full name:", first_name, last_name)
````

````python
# Creating a greeting
city = input("What city are you from? ")
print("Welcome from", city, "!")
````

---

## Critical: input() Always Returns a String

No matter what the user types, **`input()` always returns a string**:

- User types: `25` â†’ Python receives: `"25"` (string)
- User types: `3.14` â†’ Python receives: `"3.14"` (string)
- User types: `True` â†’ Python receives: `"True"` (string)

````python
# Even if user types a number, it's stored as a string
age = input("How old are you? ")
print("You entered:", age)
print("Type:", type(age))  # <class 'str'>
````

### The Problem with String Input

If we try to do math with input values, we get errors:

````python
# This will cause an error! (uncomment to see)
# age = input("Enter your age: ")
# next_year_age = age + 1  # TypeError: can only concatenate str

# Why? Because age is "25" (string), not 25 (number)
````

## Type Casting: Converting Data Types

**Type casting** (or type conversion) means changing data from one type to another.

### Main Conversion Functions

| Function | Purpose | Example |
|----------|---------|----------|
| `int()` | Convert to integer | `int("42")` â†’ `42` |
| `float()` | Convert to float | `float("3.14")` â†’ `3.14` |
| `str()` | Convert to string | `str(42)` â†’ `"42"` |

---

## 6.5 Converting to Integer: int()

Use **`int()`** to convert strings to whole numbers:

````python
# Converting a string to integer
age_string = "25"
age_number = int(age_string)

print("Original:", age_string, type(age_string))
print("Converted:", age_number, type(age_number))
````

````python
# Now we can do math!
age = input("How old are you? ")
age = int(age)  # Convert string to integer

next_year_age = age + 1
print("Next year you will be", next_year_age)
````

````python
# Shortcut: Convert immediately when getting input
age = int(input("Enter your age: "))
birth_year = 2024 - age
print("You were born around", birth_year)
````

### âš ï¸ ValueError

If you try to convert something that's not a valid number, you'll get an error:

````python
# This will cause an error (uncomment to see)
# number = int("hello")  # ValueError: invalid literal for int()
# number = int("3.14")   # ValueError: invalid literal for int()
````

## Converting to Float: float()

Use **`float()`** to convert to decimal numbers:

````python
# Converting to float
price_string = "19.99"
price_number = float(price_string)

print("Original:", price_string, type(price_string))
print("Converted:", price_number, type(price_number))
````

````python
# Using float with input
price = float(input("Enter item price: $"))
tax = price * 0.08
total = price + tax

print(f"Price: ${price}")
print(f"Tax: ${tax:.2f}")
print(f"Total: ${total:.2f}")
````

````python
# Float can also convert integers
number = float("42")  # Works! Converts to 42.0
print(number, type(number))
````

---

## Converting to String: str()

Use **`str()`** to convert numbers to strings:

````python
# The problem: Can't concatenate strings and numbers
age = 25
# message = "I am " + age + " years old"  # TypeError!
````

````python
# Solution: Convert number to string
age = 25
message = "I am " + str(age) + " years old"
print(message)
````

````python
# Alternative: Use commas in print (no conversion needed)
age = 25
print("I am", age, "years old")  # Python handles this automatically
````

````python
# Or use f-strings (modern way - we'll learn more later)
age = 25
print(f"I am {age} years old")
````

## Practical Examples

Let's build some interactive programs:

````python
# Example 1: Simple Calculator
print("=== Simple Calculator ===")
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum_result = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print(f"Sum: {sum_result}")
print(f"Difference: {difference}")
print(f"Product: {product}")
print(f"Quotient: {quotient}")
````

````python
# Example 2: Temperature Converter
print("=== Temperature Converter ===")
fahrenheit = float(input("Enter temperature in Â°F: "))
celsius = (fahrenheit - 32) * 5 / 9
print(f"{fahrenheit}Â°F = {celsius:.1f}Â°C")
````

````python
# Example 3: Personal Info Card
print("=== Create Your Info Card ===")
name = input("Name: ")
age = int(input("Age: "))
city = input("City: ")
height = float(input("Height (in feet): "))

print("\n" + "="*30)
print("      PERSONAL INFO CARD")
print("="*30)
print(f"Name:    {name}")
print(f"Age:     {age} years old")
print(f"City:    {city}")
print(f"Height:  {height} feet")
print("="*30)
````

````python
# Example 4: Shopping Cart
print("=== Shopping Cart ===")
item_name = input("Item name: ")
price = float(input("Price per item: $"))
quantity = int(input("Quantity: "))

subtotal = price * quantity
tax = subtotal * 0.08
total = subtotal + tax

print(f"\nItem: {item_name}")
print(f"Price: ${price:.2f} x {quantity}")
print(f"Subtotal: ${subtotal:.2f}")
print(f"Tax (8%): ${tax:.2f}")
print(f"Total: ${total:.2f}")
````

## Common Mistakes and How to Avoid Them

### Mistake 1: Forgetting to Convert

````python
# Wrong - will cause an error
# age = input("Age: ")
# next_year = age + 1  # TypeError

# Right - convert first
age = int(input("Age: "))
next_year = age + 1
print(next_year)
````

### Mistake 2: Using int() for Decimals

````python
# Wrong - will cause ValueError
# price = int(input("Price: "))  # If user enters 19.99, error!

# Right - use float for decimals
price = float(input("Price: "))
print(price)
````

### Mistake 3: Converting When Not Needed

````python
# Not needed - already a string, used as string
name = input("Name: ")
print("Hello", name)  # No conversion needed!

# Only convert when you need to do calculations
````

## Quick Reference

### When to Use Each Type

**Use `int()`** when:
- You need whole numbers
- Counting items, ages, years
- No decimals needed

**Use `float()`** when:
- You need decimal numbers
- Money, measurements, percentages
- Scientific calculations

**Use `str()`** when:
- You need to combine numbers with text
- Building messages
- Formatting output

**No conversion needed** when:
- Input is already text (names, addresses)
- Using commas in print
- Using f-strings
