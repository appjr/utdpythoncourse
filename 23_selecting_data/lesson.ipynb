# Topic 23: Selecting Data

## Learning Objectives
- Select columns from DataFrame
- Select rows using loc and iloc

---

## 23.1 Selecting Columns

```python
import pandas as pd

df = pd.read_csv('data.csv')
print(df['Name'])
```

## 23.2 Selecting Rows with loc

```python
# Select by label
print(df.loc[0])
```

## 23.3 Selecting with iloc

```python
# Select by position
print(df.iloc[0])
```

---

## Key Takeaways

✅ Use `df['column']` to select columns

✅ Use `.loc[]` for label-based selection

✅ Use `.iloc[]` for position-based selection
