## 16.1 What are Functions?

A **function** is a reusable block of code that performs a specific task.

### Why Use Functions?
- Avoid code repetition (DRY)
- Organize code
- Easy to test and debug
- Reusable across projects

---

## Defining a Function

```python
def function_name():
    # code block
    pass
```

````python
# Simple function
def greet():
    print("Hello!")

# Call the function
greet()
greet()  # Can call multiple times
````

````python
# Function with multiple statements
def welcome():
    print("Welcome to Python!")
    print("Let's learn functions!")
    print("Have fun coding!")

welcome()
````

## Functions with Parameters

````python
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")
greet("Bob")
````

````python
# Multiple parameters
def introduce(name, age):
    print(f"My name is {name} and I'm {age} years old.")

introduce("John", 25)
````

## Return Values

````python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)  # 8
````

## Key Takeaways
