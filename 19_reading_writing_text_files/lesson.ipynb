# Topic 19: Reading & Writing Text Files

## Learning Objectives
- Open and close files
- Read from text files
- Write to text files
- Use context managers (with statement)

---

## 19.1 Opening Files

Use `open()` to work with files:

```python
file = open('filename.txt', 'mode')
```

### Modes
- `'r'` - Read (default)
- `'w'` - Write (overwrites)
- `'a'` - Append
- `'r+'` - Read and write

## 19.2 Reading Files

````python
# Read entire file
file = open('example.txt', 'r')
content = file.read()
print(content)
file.close()
````

````python
# Read line by line
file = open('example.txt', 'r')
for line in file:
    print(line.strip())
file.close()
````

## 19.3 Writing Files

````python
# Write to file
file = open('output.txt', 'w')
file.write('Hello, World!\n')
file.write('This is a new line.\n')
file.close()

print('File written successfully!')
````

## 19.4 Appending to Files

````python
# Append to existing file
file = open('output.txt', 'a')
file.write('Appending this line.\n')
file.close()

print('Content appended!')
````

## 19.5 Using Context Managers

The `with` statement automatically closes files:

````python
# Better way - file closes automatically
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)

# File is automatically closed here
````

````python
# Writing with context manager
with open('output.txt', 'w') as file:
    file.write('Line 1\n')
    file.write('Line 2\n')
    file.write('Line 3\n')

print('File written with context manager!')
````

## 19.6 Reading Lines as List

````python
# Read all lines into a list
with open('example.txt', 'r') as file:
    lines = file.readlines()

for line in lines:
    print(line.strip())
````

---

## Key Takeaways

✅ Use `open()` to work with files

✅ Always close files with `.close()` or use `with`

✅ `'r'` for reading, `'w'` for writing, `'a'` for appending

✅ `with` statement is best practice

✅ `.read()` reads entire file, `.readlines()` returns list
