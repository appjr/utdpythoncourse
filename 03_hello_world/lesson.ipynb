## 3.2 Understanding the print() Function

The `print()` function is one of the most commonly used functions in Python. It displays output to the screen.

### Basic Syntax
```python
print(value)
```

- **print** - The function name
- **()** - Parentheses that contain what you want to print
- **"Hello, World!"** - A string (text) enclosed in quotes

### Let's Try Different Examples:

````python
# Printing a simple message
print("Welcome to Python programming!")
````

````python
# Printing numbers
print(42)
print(3.14159)
````

````python
# Printing multiple items (separated by commas)
print("Python", "is", "awesome!")
````

````python
# Printing multiple lines
print("Line 1")
print("Line 2")
print("Line 3")
````

### Single Quotes vs Double Quotes

Python allows you to use either single (`'`) or double (`"`) quotes for strings. They work the same way:

````python
# Both of these work identically
print("Hello with double quotes")
print('Hello with single quotes')
````

**When to use which?**
- Use double quotes when your text contains an apostrophe: `"I'm learning Python"`
- Use single quotes when your text contains double quotes: `'She said "Hello"'`
- Be consistent - pick one style and stick with it

````python
# Examples of when to use each type
print("I'm a Python programmer!")  # Double quotes allow apostrophe
print('The book is called "Python Basics"')  # Single quotes allow double quotes inside
````

## 3.3 Comments and Documentation

**Comments** are notes in your code that Python ignores. They're for humans, not computers!

### Why Use Comments?
- Explain what your code does
- Make notes for yourself or other programmers
- Temporarily disable code without deleting it
- Document complex logic

### Single-Line Comments

Use the `#` symbol to create a comment:

````python
# This is a comment - Python will ignore this line
print("This code runs")  # This is also a comment, after the code
````

````python
# Calculate the area of a rectangle
# Length = 5, Width = 3
print(5 * 3)
````

### Multi-Line Comments

For longer comments, you can use triple quotes (`'''` or `"""`):

````python
"""
This is a multi-line comment.
You can write as many lines as you want.
Python will ignore all of this.
"""
print("This code will run")
````

### Best Practices for Comments

âœ… **Good Comments:**

````python
# Calculate the total price with 8% tax
price = 100
tax = price * 0.08
total = price + tax
print(total)
````

âŒ **Bad Comments (too obvious):**

````python
# Print hello
print("hello")  # This comment doesn't add any value
````

## 3.4 Python Syntax Rules

Python has specific rules you must follow. Let's learn the basics:

### Rule 1: Case Sensitivity

Python distinguishes between uppercase and lowercase letters:

````python
# These are all different:
print("This works")
# Print("This will cause an error")  # Wrong - capital P
# PRINT("This will also cause an error")  # Wrong - all capitals
````

### Rule 2: Spelling Matters

You must spell function names correctly:

````python
# Correct
print("Hello")

# Wrong (uncomment to see the error)
# pring("Hello")  # Typo: pring instead of print
# pritn("Hello")  # Typo: pritn instead of print
````

### Rule 3: Matching Quotes

Opening and closing quotes must match:

````python
# Correct - matching quotes
print("Hello")
print('Hello')

# Wrong - mismatched quotes (uncomment to see error)
# print("Hello')  # Error: starts with " but ends with '
# print('Hello")  # Error: starts with ' but ends with "
````

### Rule 4: Parentheses Must Balance

Every opening parenthesis needs a closing one:

````python
# Correct
print("Hello")

# Wrong (uncomment to see error)
# print("Hello"  # Missing closing parenthesis
# print"Hello")  # Missing opening parenthesis
````

## The Importance of Indentation

**Indentation** (spaces or tabs at the beginning of a line) is CRITICAL in Python.

### Why Indentation Matters

Unlike many other languages that use braces `{}` or keywords, Python uses indentation to define code blocks. This makes Python code clean and readable!

### Preview: Indentation in Action

We'll learn more about this later, but here's a sneak peek:

````python
# Example 1: No indentation needed (for now)
print("This is not indented")
print("Neither is this")
````

````python
# Example 2: Preview of indentation (we'll learn this later)
# Don't worry about understanding this yet - just observe the indentation
x = 10
if x > 5:
    print("This line is indented")  # This line MUST be indented
    print("So is this one")  # Same level of indentation
````

### Indentation Rules

1. **Use 4 spaces** (Python convention) - Jupyter does this automatically!
2. **Be consistent** - Don't mix tabs and spaces
3. **All code in a block** must have the same indentation

**Good News**: For simple programs like those we're writing now, you don't need to worry about indentation yet. It becomes important when we learn about conditionals and loops!

## Understanding Error Messages

When you make a mistake, Python gives you an **error message**. Don't be afraid of errors - they're helping you fix the problem!

### Example: Syntax Error

````python
# Uncomment the line below to see a syntax error
# print("Hello)

# The error message tells you:
# 1. What type of error (SyntaxError)
# 2. Where the error is (line number)
# 3. What's wrong (unterminated string literal)
````

### Common Error Types

**SyntaxError**: You wrote something Python doesn't understand
- Missing quotes, parentheses, or colons
- Misspelled keywords

**NameError**: You used a name Python doesn't recognize
- Typo in function name
- Using a variable before defining it

**IndentationError**: Your indentation is incorrect
- Mixed tabs and spaces
- Unexpected indent

### How to Read Error Messages

1. **Look at the error type** (last line)
2. **Read the message** - it usually explains the problem
3. **Check the line number** - shows where the error occurred
4. **Look at the code** - Python shows you the problematic line

## More print() Function Features

Let's explore more capabilities of the `print()` function:

### Printing Multiple Items

````python
# Separate items with commas
print("Python", "is", "fun!")
````

````python
# Mix text and numbers
print("I am", 25, "years old")
````

### Changing the Separator

````python
# Default separator is a space
print("Hello", "World")

# Use sep parameter to change it
print("Hello", "World", sep="-")
print("Hello", "World", sep="***")
print("2024", "12", "08", sep="/")
````

### Changing the End Character

````python
# By default, print() adds a newline at the end
print("Line 1")
print("Line 2")

# Use end parameter to change it
print("Same", end=" ")
print("line!")

print("With", end="***")
print("stars!")
````

### Printing Empty Lines

````python
print("First line\n")
print()  # Empty print adds a blank line
print("Third line (with a blank line above)")
````

## Escape Characters

**Escape characters** let you include special characters in strings. They start with a backslash `\`.

### Common Escape Characters

- `\n` - New line (line break)
- `\t` - Tab (indent)
- `\\` - Backslash
- `\'` - Single quote
- `\"` - Double quote

````python
# New line
print("Line 1\nLine 2\nLine 3")
````

````python
# Tab spacing
print("Name:\tJohn")
print("Age:\t25")
print("City:\tNew York")
````

````python
# Quotes inside strings
print("She said, \"Hello!\"")
print('It\'s a beautiful day')
````

````python
# Backslash
print("C:\\Users\\Documents")
````
