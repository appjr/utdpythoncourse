# Topic 15: Introduction to Dictionaries

## Learning Objectives
- Understand dictionaries as key-value pairs
- Create and modify dictionaries
- Access dictionary values
- Use dictionary methods

---

## 15.1 What are Dictionaries?

A **dictionary** stores data as key-value pairs.

```python
person = {
    "name": "Alice",
    "age": 25,
    "city": "Dallas"
}
```

## 15.2 Creating Dictionaries

````python
# Empty dictionary
empty = {}

# Dictionary with data
student = {
    "name": "John",
    "age": 20,
    "major": "Computer Science"
}
print(student)
````

## 15.3 Accessing Values

````python
student = {"name": "John", "age": 20}

print(student["name"])  # John
print(student.get("age"))  # 20
````

## 15.4 Modifying Dictionaries

````python
student = {"name": "John", "age": 20}

# Update value
student["age"] = 21

# Add new key-value
student["gpa"] = 3.8

print(student)
````

## 15.5 Dictionary Methods

````python
student = {"name": "John", "age": 20, "major": "CS"}

print(student.keys())    # All keys
print(student.values())  # All values
print(student.items())   # All key-value pairs
````

## Iterating Over Dictionaries

````python
student = {"name": "John", "age": 20, "major": "CS"}

for key, value in student.items():
    print(f"{key}: {value}")
````

---

## Key Takeaways

✅ Dictionaries use key-value pairs

✅ Access with dict[key] or dict.get(key)

✅ Keys must be unique

✅ Mutable - can add/change/remove items
