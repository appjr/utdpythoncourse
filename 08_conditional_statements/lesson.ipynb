## Introduction to Decision Making

Real programs need to make decisions:
- If password is correct, log in

**Conditional statements** let your program choose different paths based on conditions.

---

## 8.2 The if Statement

The **if statement** runs code only if a condition is True.

### Syntax
```python
if condition:
    # code to run if True
    # must be indented!
```

````python
# Simple if statement
age = 20

if age >= 18:
    print("You are an adult")
    print("You can vote")
````

````python
# With user input
temperature = float(input("Temperature: "))

if temperature > 90:
    print("It's very hot!")
    print("Stay hydrated")
````

## 8.3 The else Statement

The **else statement** runs when the if condition is False.

````python
# if-else example
age = 15

if age >= 18:
    print("You can vote")
else:
    print("You cannot vote yet")
````

````python
# Password checker
password = input("Enter password: ")

if password == "python123":
    print("Access granted")
else:
    print("Access denied")
````

## 8.4 The elif Statement

The **elif statement** (else-if) checks additional conditions.

````python
# Grade classifier
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
````

````python
# Temperature categories
temp = float(input("Temperature: "))

if temp >= 90:
    print("Very hot")
elif temp >= 70:
    print("Warm")
elif temp >= 50:
    print("Cool")
else:
    print("Cold")
````

## Nested if Statements

You can put if statements inside other if statements.

````python
age = 20
has_license = True

if age >= 16:
    if has_license:
        print("You can drive")
    else:
        print("You need a license")
else:
    print("You're too young to drive")
````

## 8.8 Practical Examples

````python
# Example 1: Even or Odd
number = int(input("Enter a number: "))

if number % 2 == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")
````

````python
# Example 2: Positive, Negative, or Zero
num = float(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
````

````python
# Example 3: Ticket Price Calculator
age = int(input("Enter your age: "))

if age < 3:
    price = 0
    print("Admission is free!")
elif age < 13:
    price = 10
    print(f"Child ticket: ${price}")
elif age < 65:
    price = 15
    print(f"Adult ticket: ${price}")
else:
    price = 10
    print(f"Senior ticket: ${price}")
````

````python
# Example 4: BMI Calculator with categories
weight = float(input("Weight (kg): "))
height = float(input("Height (m): "))

bmi = weight / (height ** 2)
print(f"Your BMI: {bmi:.1f}")

if bmi < 18.5:
    print("Category: Underweight")
elif bmi < 25:
    print("Category: Normal weight")
elif bmi < 30:
    print("Category: Overweight")
else:
    print("Category: Obese")
````
