# Topic 11: Loop Control - break and continue

## Learning Objectives
- Use break to exit loops early
- Use continue to skip iterations
- Control loop flow for complex logic

---

## The break Statement

**break** exits the loop immediately.

```python
# Stop when finding target
for num in range(1, 11):
    if num == 5:
        print("Found 5! Breaking...")
        break
    print(num)
```

```python
# Search in a list
names = ["Alice", "Bob", "Charlie", "David"]
search = "Charlie"

for name in names:
    if name == search:
        print(f"Found {search}!")
        break
else:
    print(f"{search} not found")
```

## The continue Statement

**continue** skips the current iteration and moves to the next.

```python
# Skip odd numbers
for num in range(1, 11):
    if num % 2 != 0:
        continue  # Skip odd numbers
    print(num)  # Only even numbers print
```

```python
# Skip negative numbers
numbers = [5, -2, 8, -7, 3, -1, 10]

for num in numbers:
    if num < 0:
        continue  # Skip negative
    print(num)  # Only positive numbers
```

## Practical Example

```python
# Password validator
max_attempts = 3
correct_password = "python123"

for attempt in range(1, max_attempts + 1):
    password = input(f"Attempt {attempt}: Enter password: ")
    
    if password == correct_password:
        print("Access granted!")
        break
    else:
        print("Wrong password")
        if attempt < max_attempts:
            continue
else:
    print("Locked out!")
```
