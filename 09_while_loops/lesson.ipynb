## What are Loops?

**Loops** repeat code multiple times automatically.

Without loops:
```python
print("Hello")
print("Hello")
print("Hello")
```

With loops:
```python
for i in range(3):
    print("Hello")
```

---

## The while Loop

A **while loop** repeats as long as a condition is True.

### Syntax
```python
while condition:
    # code to repeat
```

````python
# Count from 1 to 5
count = 1
while count <= 5:
    print(count)
    count += 1  # Important: update the counter!
````

````python
# Countdown
n = 5
while n > 0:
    print(n)
    n -= 1
print("Blast off!")
````

## Avoiding Infinite Loops

**Critical**: Always update the loop variable!

```python
# INFINITE LOOP - Don't run!
count = 1
while count <= 5:
    print(count)
    # Missing: count += 1
```

## Practical Examples

````python
# Password checker with retry
correct_password = "python123"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
    password = input("Enter password: ")
    if password == correct_password:
        print("Access granted!")
        break
    else:
        attempts += 1
        print(f"Wrong! {max_attempts - attempts} attempts left")

if attempts == max_attempts:
    print("Locked out!")
````
