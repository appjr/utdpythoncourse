## Learning Objectives

---

## 24.1 Filtering Data

```python
import pandas as pd

df = pd.read_csv('data.csv')
filtered = df[df['Age'] > 25]
print(filtered)
```

## 24.2 Multiple Conditions

```python
# Filter with multiple conditions
filtered = df[(df['Age'] > 25) & (df['City'] == 'NYC')]
print(filtered)
```

## 24.3 Sorting Data

```python
# Sort by column
sorted_df = df.sort_values('Age')
print(sorted_df)
```
