## 18.1 What are Modules?

A **module** is a file containing Python code (functions, variables, classes).

### Why Use Modules?
- **Reusability**: Use code written by others
- **Organization**: Keep code organized
- **Efficiency**: Don't reinvent the wheel

## 18.2 Importing Modules

Use `import` to access module functionality:

```python
# Import the math module
import math

# Use functions from math
result = math.sqrt(16)
print(result)  # 4.0
```

## 18.3 The math Module

```python
import math

print(math.sqrt(25))  # 5.0
print(math.pow(2, 3))  # 8.0
print(math.pi)  # 3.141592653589793
```

## 18.4 The random Module

```python
import random

# Random integer
number = random.randint(1, 10)
print(number)

# Random choice
colors = ["red", "blue", "green"]
print(random.choice(colors))
```

## 18.5 Importing Specific Functions

```python
# Import only sqrt from math
from math import sqrt, pi

print(sqrt(25))  # 5.0
print(pi)  # 3.141592653589793
```

## 18.6 Module Aliases

```python
# Use alias for shorter names
import math as m

print(m.sqrt(16))  # 4.0
print(m.pi)  # 3.141592653589793
```

---

## Key Takeaways

✅ Modules are reusable Python code files

✅ Use `import` to access modules

✅ `from module import function` for specific items

✅ `import module as alias` for shorter names
