## 10.1 The for Loop

A **for loop** iterates over a sequence.

### Syntax
```python
for item in sequence:
    # code to repeat
```

````python
# Iterate over a string
for letter in "Hello":
    print(letter)
````

## 10.2 The range() Function

**range()** generates a sequence of numbers.

- `range(5)` â†’ 0, 1, 2, 3, 4
- `range(1, 6)` â†’ 1, 2, 3, 4, 5
- `range(0, 10, 2)` â†’ 0, 2, 4, 6, 8

````python
# Count from 1 to 5
for i in range(1, 6):
    print(i)
````

````python
# Even numbers
for n in range(0, 11, 2):
    print(n)
````

## 10.3 Practical Examples

````python
# Multiplication table
number = 5
for i in range(1, 11):
    print(f"{number} x {i} = {number * i}")
````

````python
# Sum of numbers
total = 0
for num in range(1, 101):
    total += num
print(f"Sum: {total}")
````
