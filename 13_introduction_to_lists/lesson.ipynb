## 13.1 What are Lists?

A **list** is an ordered, mutable collection of items.

```python
fruits = ["apple", "banana", "orange"]
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]
```

## 13.2 Creating Lists

````python
# Empty list
empty = []

# List with items
fruits = ["apple", "banana", "orange"]
numbers = [1, 2, 3, 4, 5]

print(fruits)
print(len(fruits))  # Length of list
````

## Accessing Items

Use **indexing** (starts at 0):

````python
fruits = ["apple", "banana", "orange"]

print(fruits[0])   # First item
print(fruits[1])   # Second item
print(fruits[-1])  # Last item
````

## 13.6 Modifying Lists

````python
fruits = ["apple", "banana", "orange"]

# Change an item
fruits[1] = "grape"
print(fruits)
````

## 13.6 List Methods

### append() - Add to end

````python
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)
````

### remove() - Remove by value

````python
fruits = ["apple", "banana", "orange"]
fruits.remove("banana")
print(fruits)
````

### pop() - Remove by index

````python
fruits = ["apple", "banana", "orange"]
removed = fruits.pop(1)  # Remove index 1
print(f"Removed: {removed}")
print(f"Remaining: {fruits}")
````

## Iterating Over Lists

````python
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
````
