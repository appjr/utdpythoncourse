# Topic 17: Function Parameters & Return Values

## Learning Objectives
- Use parameters to pass data into functions
- Use return to get data out of functions
- Understand default parameters
- Work with multiple return values

---

## Parameters

**Parameters** allow functions to receive input:

```python
def calculate_area(length, width):
    area = length * width
    return area

result = calculate_area(5, 3)
print(result)  # 15
```

## Return Statement

**return** sends data back from a function:

```python
def square(n):
    return n * n

result = square(5)
print(result)  # 25
```

## 17.6 Default Parameters

```python
def greet(name, message="Hello"):
    return f"{message}, {name}!"

print(greet("Alice"))  # Hello, Alice!
print(greet("Bob", "Hi"))  # Hi, Bob!
```

## 17.5 Multiple Return Values

```python
def calculate_stats(numbers):
    total = sum(numbers)
    average = total / len(numbers)
    return total, average

t, avg = calculate_stats([1, 2, 3, 4, 5])
print(f"Total: {t}, Average: {avg}")
```

---

## Key Takeaways

✅ Parameters pass data INTO functions

✅ return sends data OUT of functions

✅ Default parameters provide fallback values

✅ Functions can return multiple values
